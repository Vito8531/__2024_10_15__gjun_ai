```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)
objects = ["chinese","english","math","social","science"]
score = [76,70,100,84,94]
score1 = [80,62,94,88,64]
ax_1.plot(objects,score,"ro-")
ax_1.plot(objects,score1,"g*--")
ax_1.set_title("students")
ax_1.set_xlabel("objects")
ax_1.set_ylabel("score")
ax_1.grid(axis="y")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1, 1, 1)
subjects = ['chinese','english','math','history','discover']
nums1 = [1, 4, 9, 16, 25]
nums2 = [25, 16, 12, 4, 1]
axes.plot(subjects,nums1,'+--y',label='student1')
axes.plot(subjects,nums2,'o-r',label='student2')
axes.set_title('students')
axes.set_xlabel('subject')
axes.set_ylabel('scores')
axes.grid(axis='y')
axes.set_xticks([0,1,2,3,4])
axes.set_xticklabels(['ch','en','ma','hi','cover'])
axes.legend()
plt.show()
```
